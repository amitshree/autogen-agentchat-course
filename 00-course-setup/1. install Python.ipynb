# Installing Python 3.12

This guide provides instructions for installing Python 3.12 on Windows, macOS, and Linux.

## Windows

1. **Download the installer:**

   * Visit the official Python website: [https://www.python.org/downloads/windows/](https://www.python.org/downloads/windows/)
   * Download the appropriate *Windows installer (64-bit)*.  32-bit is extremely rare these days.

2. **Run the installer:**

   * Double-click the downloaded executable file.
   * **Crucial:** Ensure you check the box that says "Add Python 3.12 to PATH" during the installation process. This is essential for running Python from the command prompt or PowerShell.
   * You can choose to "Install Now" for a standard installation, or "Customize installation" if you need more control.

3. **Verify installation:**

   * Open the command prompt (search for "cmd" in the Start Menu) or PowerShell.
   * Type `python --version` and press Enter. You should see output similar to `Python 3.12.x`.

## macOS

macOS already comes with Python, but it's typically an older version and *not* recommended for development. We'll install a newer version alongside it.

1. **Install Homebrew (if not already installed):**

   * Homebrew is a package manager that simplifies software installation. If you don't have it, open Terminal and paste the following command:

     ```bash
     /bin/bash -c "$(curl -fsSL [https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh](https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh))"
     ```

2. **Install Python 3.12 using Homebrew:**

   * In Terminal, run:

     ```bash
     brew install python@3.12
     ```

3. **Verify installation:**

   * Open Terminal and type `python3.12 --version`. You should see `Python 3.12.x`.  Crucially, you'll use `python3.12` to invoke this version.  

4. **Set Python 3.12 as the default (Recommended):**
   * Open your .bash_profile file in a text editor. If it doesn't exist, create it. `nano ~/.bash_profile`  # Or use your preferred editor (e.g., vim, emacs) and add these 2 lines
   
   ```bash
   alias python=python3
   alias pip=pip3
